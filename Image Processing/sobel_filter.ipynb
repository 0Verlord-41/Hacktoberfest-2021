```
import numpy as np
import  cv2 as cv
from matplotlib import pyplot as plt
```

```
img = cv.imread('/content/camerman.png')
```

```
plt.imshow(img)
plt.xticks([]),plt.yticks([])
plt.show()
```

```
sobelx = cv.Sobel(img,cv.CV_64F,1,0,ksize=3)
```

```
plt.subplot(1,1,1),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
```

```
sobely = cv.Sobel(img,cv.CV_64F,0,1,ksize=3)
```

```
plt.subplot(1,1,1),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
```

```
sobel_add = (sobelx+sobely)
```

```
plt.subplot(1,1,1),plt.imshow(sobel_add)
plt.title('Sobel ADD'), plt.xticks([]), plt.yticks([])
```

Note - Here x+y gives the raw magnitude but to get gradient magnitude it should be sqrt(x^2+y^2)
